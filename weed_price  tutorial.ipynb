```python
# Position of the Outlier 
# It shows that the outlier lies below 250
print(np.where(df['HighQ']<250))
```

```python
# Scatter plot for detection of outlier
fig, ax = plt.subplots(figsize=(18,10))
ax.scatter(df['Year'],df['HighQ'])
ax.set_xlabel('Year')
ax.set_ylabel('HighQ')
plt.show
```

```python
# In scatter plot shows the year wise outlier ,using appox we can say that in 2013 (outlier<250),
#  in 2014 (outlier <250), in 2015 (outlier<250 & outlier>375)

```

```python
# Z Score = This value/score helps to understand that the how far is data point from mean.
#  Zscore=(data_point-mean/std.deviation)
from scipy import stats
z =np.abs(stats.zscore(df['HighQ']))
z
```

```python
# Position of outliers
threshold = 3
print(np.where(z > 3))
```

```python
# IQR(Inter Quartile Range) approach to finding the outlier is the most commonly used 
# and most trusted approach used in research field.
# IQR = Quartile3 - Quartile1
Q1 = np.percentile(df['HighQ'], 25, interpolation = 'midpoint')
Q3 = np.percentile(df['HighQ'], 75, interpolation = 'midpoint')
IQR = Q3 - Q1
print('IQR:',IQR,'Q1:',Q1,'Q3:',Q3)
```
